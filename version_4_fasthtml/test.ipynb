```python
from fasthtml.common import *
import pandas as pd

# Load and merge data
basic_info_df = pd.read_csv('basic_info.csv')
op_status_df = pd.read_csv('op_status.csv')
merged_df = pd.merge(basic_info_df, op_status_df, on=['name', 'id', 'macAddress'])
```

```python

app, rt = fast_app()

# Function to generate the health indicator color
def health_indicator(status):
    if status == 100:
        return 'green'
    elif 50 <= status < 100:
        return 'yellow'
    else:
        return 'red'
```

```python

# Homepage with health indicators
@rt('/')
def homepage():
    devices = []
    for _, row in merged_df.iterrows():
        color = health_indicator(row['connectivityStatus'])
        devices.append(
            Div(
                A(row['name'], href=f'/device/{row["name"]}', cls='device-link'),
                Span(cls=f'health-indicator {color}')
            )
        )
    return Div(*devices)

# # Route for detailed device info
# @rt('/device/{name}')
# def device_detail(request, name):
#     try:
#         # Find the device by name
#         device_info = merged_df[merged_df['name'] == name]
        
#         if device_info.empty:
#             return Div(H2(f'Device "{name}" not found'), A('Back to Home', href='/'))
        
#         device_info = device_info.iloc[0]

#         return Div(
#             H2(f'Device: {device_info["name"]}'),
#             P(f'ID: {device_info["id"]}'),
#             P(f'MAC Address: {device_info["macAddress"]}'),
#             P(f'Type: {device_info["apType"]}'),
#             P(f'Location: {device_info["location"]}'),
#             P(f'Uptime: {device_info["upTime"]} seconds'),
#             P(f'Connected Time: {device_info["connectedTime"]} seconds'),
#             A('Back to Home', href='/')
#         )
#     except Exception as e:
#         return Div(H2('Error: Could not retrieve device information'), P(str(e)), A('Back to Home', href='/'))

@rt('/device/{name}')
def device_detail(request, name):
    # 'name' contains the device name from the URL
    return Div(H2(f'Device: {name}'))

serve()
```
